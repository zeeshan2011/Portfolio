# web scrapping

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('div')
```

```python
soup.find_all('div')
```

```python
soup.find_all('div', class_ = "col-md-12")
#specifying the div with the class = "col-md-12"
```

```python
soup.find('p')
```

```python
soup.find_all('p')
```

```python
soup.find_all('p', class_ = 'lead')
```

```python
#cleaning a little bit
soup.find('p', class_ = 'lead').text.strip()
```

```python
soup.find_all('tr')
```

```python
soup.find_all('tr', class_team_name = 'Boston Bruins')
```
