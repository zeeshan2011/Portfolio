# Merging , Join & Concatenate

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\pshan\Downloads\LOTR.csv")
```

```python
df2 = pd.read_csv(r"C:\Users\pshan\Downloads\LOTR 2.csv")
```

```python
df2
```

```python
df.merge(df2)
#by default us the inner join 
# to specify use how 
```

```python
df.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
df.merge(df2, how = 'left')
```

```python
df.merge(df2, how = 'cross')
```

```python
# joins 
df3 = df.set_index('FellowshipID')
df4 = df2.set_index('FellowshipID')
```

```python
df3.join(df4, lsuffix = 'Left', rsuffix = 'Right', how = 'inner')
```

```python
#concat put one database on top of the other
# merge and join put them next to eachother
```

```python
pd.concat([df3, df4])
```
