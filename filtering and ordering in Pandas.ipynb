# Filtering and Ordering in Pandas

```python
df_asia = df[ df['Continent'] == 'Asia']
df_europe = df[ df['Continent'] == 'Europe']
```

```python
df_europe[df_europe['Rank'] > 50]
```

```python
specific_countries = ['Brazil', 'Bangladesh']
df[df['Country'].isin(specific_countries)]
```

```python
# to set an index
df2 = df.set_index('Country')
```

```python
#filter based on specific columns 
df.filter(items = ['Rank', 'Country', 'Continent', '2022 Population'], axis = 1)
#axis = 0 -> search on the x-axis and axis = 1 -> search on the y-axis
```
